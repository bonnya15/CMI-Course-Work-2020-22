```python
def insertionSort(l):
    for i in range(1,len(l)):
        key = l[i]
        j = i-1
        while j > 0 and l[j+1] > l[j]:
            l[j+1] = l[j]
            j -= 1
        l[j+1] = key 
    return(l)
```

```python
insertionSort(l)
```

```python
def addpoly(p1,p2):
    l1 = []
    l2 = []
    empty_dict = dict()
    for keys in p1.keys():
        l1.append(keys)
    for k in p2.keys():
        l2. append(k)
    for l in l1:
        if l in l2 and p1[l] + p2[l] !=0:
            empty_dict[l] = p1[l] + p2[l]
        elif l not in l2:
            empty_dict[l] = p1[l]
    for k in l2:
        if k not in l1:
            empty_dict[k] = p2[k]
    return(empty_dict)
            
        
```
