```python
def factorial(n):
   if (n == 0):
      return(1)
   else:
      return(n * factorial(n-1))
```

```python
factorial(-1)
```

```python
factorial(20)
```

```python
def mylen(l):
    if l == []:
        return(0)
    else:
        return(1+mylen(l[1:]))
```

```python
mylen(list(range(1000)))
```

```python
def mysum(l):
    if l == []:
        return(0)
    else:
        return(l[0] + mysum(l[1:]))
```

```python
mysum(list(range(10)))
```

```python
mysum(['the','long','road'])
```

```python
def zigzag(l):
    return(updown(l) or downup(l))

def updown(l):
    if len(l) < 2:
        return(True)
    else:
        return(l[0] < l[1] and downup(l[1:]))
    
def downup(l):
    if len(l) < 2:
        return(True)
    else:
        return(l[0] > l[1] and updown(l[1:]))
```

```python
zigzag([0,1,0,1,0])
```

```python
zigzag([1,0,1,0,1])
```
